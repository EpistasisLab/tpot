`conda-forge` is growing organization on Github around the Anaconda ecosystem.  

> [`conda-forge`](conda-forge.github.io) is a github organization containing repositories of conda recipes. Thanks to some awesome continuous integration providers (AppVeyor, CircleCI and TravisCI), each repository, also known as a feedstock, automatically builds its own recipe in a clean and repeatable way on Windows, Linux and OSX.

```python
p=figure(x_axis_type='datetime', height=400, width=800)
source = releases.groupby(releases.index).agg({
        'ndownloads': 'sum', 'upload_time': 'min'
    }).sort_values('upload_time').reset_index()
source['cdownloads'] = source.ndownloads.cumsum(); source = source.pipe(ColumnDataSource)

p.add_tools(HoverTool(
        renderers=[p.line(source=source,x='upload_time',y='cdownloads')], tooltips="""@name - @ndownloads""",
    ))

show(p)
```
